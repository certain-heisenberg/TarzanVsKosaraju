```python
import os
import glob
import sys
```

```python
base_dir = os.getcwd()
# print(base_dir)

kosaraju = open("kosaraju.txt", "r")
tarzan = open("tarzan.txt", "r")
```

```python
vertices = []
edges = []
scc_cnt = []

kosaraju_runtime = []

for line in kosaraju:
    content = line.split(' ')
    vertices.append(int(content[0]))
    edges.append(int(content[1]))
    scc_cnt.append(int(content[2]))
    kosaraju_runtime.append(int(content[3]))
```

```python
tarzan_runtime = []

for line in tarzan:
    content = line.split(' ')
    tarzan_runtime.append(int(content[3]))
```

```python
len(vertices)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_v = np.array(vertices)
y_k = np.array(kosaraju_runtime)
y_t = np.array(tarzan_runtime)

plt.scatter(x_v, y_k)
plt.scatter(x_v, y_t)
plt.xlabel('Number of vertices')
plt.ylabel('Running time (in microseconds)')
plt.title('Runtime vs Vertices Count')
plt.legend(['Kosaraju', 'Tarzan'])
plt.show()
```

```python
x_e = np.array(edges)
y_k = np.array(kosaraju_runtime)
y_t = np.array(tarzan_runtime)

plt.scatter(x_e, y_k)
plt.scatter(x_e, y_t)
plt.xlabel('Number of edges')
plt.ylabel('Running time (in microseconds)')
plt.title('Runtime vs Edges Count')
plt.legend(['Kosaraju', 'Tarzan'])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

X = np.array(vertices)
Y = np.array(edges)
X, Y = np.meshgrid(X, Y)
size = len(kosaraju_runtime)
Z = np.array(kosaraju_runtime).reshape(size/2,2).round().astype(np.int32)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
